Here, we will use an insurance dataset to predit the premium to be paid by a person with 100yo 

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# TO DO: load dataset
df = pd.read_csv('../datasets/insurance.csv')
# TO DO: identify variables (independent and dependent)
x = df['Age'].values.reshape(-1,1)
y = df['Premium']
# TO DO: train the linear regression model
model = LinearRegression()
model.fit(x, y)
# TO DO: make predictions
y_predicted = model.predict(x)
# TO DO: visualize the data and model
plt.scatter(x, y)
plt.plot(x, y_predicted)
plt.title('Age vs Premium')
plt.xlabel('Age')
plt.ylabel('Premium')
# TO DO: predict the premium for a 100 year old person
age_100 = np.array([[100]])
premium_predicted_100 = model.predict(age_100)
print('the premium to be paid by a person with 100years is: ',premium_predicted_100[0])
```
